```python
import matplotlib.pyplot as plt 
import random
import math
```

```python
rms=[]       # array for storing the rms values for each set N

N=[]         # array of square root of step numbers
for n in range(250,651,100):     # N=250,350....650
    r=[]     # array for radial distance for each trial in N
    print("No of steps N=",n)
    for i in range(100):       # 100 trials for each N
        # intialising
        x = [0.0 for j in range(n+1)]  # array for random x
        y = [0.0 for k in range(n+1)]  # array for random y
        pi=math.pi    # pi value
        sumx,sumy,xd,yd=0.0,0.0,0.0,0.0
        for l in range(n):    # n steps
            theta=random.uniform(0,2*pi)  # random value of theta in [0,2pi]
            x[l+1]=x[l]+math.sin(theta)   # next x value
            y[l+1]=y[l]+math.cos(theta)   # next y value
        for k in range(n):
            sumx+=(x[k+1]-x[k])   # sum of del(x)
            sumy+=(y[k+1]-y[k])   # sum of del(y)
        # finding average displacement in x and y axes (mean)
        xd=sumx/n
        yd=sumy/n

        p=math.sqrt(((x[n]-x[0])**2)+((y[n]-y[0])**2)) # radial distance for ith trial
        print("For ", i+1, "th trial, radial distance =", p)
        print("Average displacement in X direction=", xd)
        print("Average displacement in Y direction=", yd)
        print()
        r.append(p)
        if i<5:          # only plotting 5 trials in a graph (for each N)
            plt.plot(x,y,label="Trial no.=%s"%(i+1))  # plot random walk
            plt.title('N=%i'%n)
    plt.legend()
    plt.show()
    sum=0.0
    for c in range(len(r)):
        sum+=r[c]**2      # sum of square of radial distances
    # square root of mean of the sum= R_rms
    meansq=math.sqrt(sum/len(r))
    print("Value of R_rms=",meansq)
    print()
    print()
    rms.append(meansq)     # storing the rms values for each N
    N.append(math.sqrt(n))            # sqrt of N

# sqrt(n) vs R_rms plot
plt.plot(N,rms,label='R_rms values',marker=".",markersize=5)
plt.plot(rms,rms,'--',label="Analytical curve")
plt.legend()
plt.xlabel("sqrt(N)")
plt.ylabel("R_rms")
plt.title("R_rms vs sqrt(N)")
plt.show()
```
